```python
class Ship(object):
    def __init__(self, damage):
        self.damage = damage
    
    def deal_damage(self):
        print("{} dealt damage of {}".format(self.__class__.__name__, self.damage))
```

```python
class BattleShip(Ship):
    def __init__(self, damage, special_damage):
        super(BattleShip, self).__init__(damage)
        self.special_damage = special_damage
    
    def deal_special_damage(self):
        print("{} dealt extra damage of {}".format(self.__class__.__name__, self.special_damage))
```

```python
class BigBattleShip(BattleShip):
    def __init__(self, damage, special_damage, bomb_damage):
        super(BigBattleShip, self).__init__(damage, special_damage)
        self.bomb_damage = bomb_damage
        
    def deal_special_damage_twice(self):
        super(BigBattleShip, self).deal_special_damage()
        super(BigBattleShip, self).deal_special_damage()
    
    def use_bomb(self):
        print("{} dealt extra damage of {} by using Bomb".format(self.__class__.__name__, self.special_damage))
```

```python
class CargoShip(Ship):
    def __init__(self, damage, capacity):
        super(CargoShip, self).__init__(damage)
        self.capacity = capacity
        
    def transport(self):
        print("{} has capacity of {} for Transportation".format(self.__class__.__name__, self.capacity))
```

```python
if __name__ == "__main__":
    ship = Ship(25)
    battle = BattleShip(20, 50)
    big = BigBattleShip(15, 60, 100)
    cargo = CargoShip(10, 200)
    
    print("Info for Ship:")
    ship.deal_damage()
    
    print()
    
    print("Info for Battle Ship:")
    battle.deal_damage()
    battle.deal_special_damage()
    
    print()
    
    print("Info for Big Battle Ship:")
    big.deal_damage()
    big.deal_special_damage_twice()
    big.use_bomb()
    
    print()
    
    print("Info for Cargo Ship:")
    cargo.deal_damage()
    cargo.transport()
```
